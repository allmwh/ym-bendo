```python
import requests
import pandas as pd
from pathlib import Path
from selenium import webdriver
import time
```

```python
driver_path = Path("/home/wenlin/d/custom_command/chromedriver")
base_url = 'https://portal.ym.edu.tw/umchi/activity/'
```

```python
def get_page_list():
    page_list = []
    a = driver.find_elements_by_tag_name('option')
    for i in a:        
        page_list.append(i.get_attribute('value'))
    del page_list[0]
    return page_list

def get_event_link(event_name):
    try:
        return driver.find_element_by_link_text(event_name).get_attribute('href')
    except:
        return 'exception'
    
def have_bendo(event_link):
    try:
        driver.get(event_link)
        checked = driver.find_element_by_xpath("//input[@value='food']").get_attribute('checked')
        if checked == 'true':
            return 1
        else:
            return 0
    except:
        return 0
    
def get_event_list(page_list):
    event_list = pd.DataFrame()
    for i in page_list:
        driver.get(base_url+i)
    
        table = pd.read_html(base_url+i,
                encoding = 'utf8',
                header=0)[2]
        
        ##網址column
        table['網址']=table['活動名稱與說明'].apply(get_event_link)
        
        ##有無便當column
        table['便當']=table['網址'].apply(have_bendo)
        
        event_list = event_list.append(table)
    event_list.reset_index(drop=True, inplace=True)
    return event_list
```

```python
options = webdriver.ChromeOptions()
options.add_argument("--headless")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
```

```python
driver = webdriver.Chrome(driver_path, chrome_options=options)
driver.get(base_url)

page_list = get_page_list()
event_list = get_event_list(page_list)
bendo_list = event_list[event_list['便當']==1]

bendo_list[['活動名稱與說明','活動日期','網址']].to_excel('便當.xlsx')
```
